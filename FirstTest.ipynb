#想要把第一個CSV檔讀入

```python
import pandas as pd

df = pd.read_csv('FirstTest.csv')

df.head()
```

```python
temp = df[['Complex2Name','FValue']]
temp
```

```python
Type = temp.groupby(['Complex2Name'])['FValue'].sum().reset_index()
```

```python
Type['year'] = 2015
Type
```

#讀入2016－2018年的預算資料

```python
df2016 = pd.read_csv('2016.csv')
df2017 = pd.read_csv('2017.csv')
df2018 = pd.read_csv('2018.csv')
```

```python
new = df2016[['Complex2Name','FValue']].groupby(['Complex2Name'])['FValue'].sum().reset_index()
```

```python
new['year'] = 2016
new
```

```python
frames = [Type,new]

result = pd.concat(frames)

result
```

```python
baby = df2017[['Complex2Name','FValue']].groupby(['Complex2Name'])['FValue'].sum().reset_index()
```

```python
baby['year'] = 2017
baby
```

```python
frames = [Type,new,baby]

result = pd.concat(frames)

result
```

```python
hello = df2018[['Complex2Name','FValue']].groupby(['Complex2Name'])['FValue'].sum().reset_index()
```

```python
frames = [Type,new,baby,hello]

result = pd.concat(frames)

result
```
