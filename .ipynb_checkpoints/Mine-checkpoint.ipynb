#109年各季度地方教育發展基金補助

```python
import pandas as pd

df = pd.read_csv('109-1.csv')

df.head() 
```

```python
temp = df[['補助事項或用途','補助計畫案總經費及分攤情形-合計']]
temp
```

```python
Type = temp.groupby(['補助事項或用途'])['補助計畫案總經費及分攤情形-合計'].sum().reset_index()
```

```python
Type['季度']=1
Type
```

```python
df2 = pd.read_csv('109-2.csv')
df3 = pd.read_csv('109-3.csv')
df4 = pd.read_csv('109-4.csv')
```

```python
new = df2[['補助事項或用途','補助計畫案總經費及分攤情形-合計']].groupby(['補助事項或用途'])['補助計畫案總經費及分攤情形-合計'].sum().reset_index()
```

```python
new['季度'] = 2
new
```

```python
frames = [Type, new]

result = pd.concat(frames)

result
```

```python
hello = df3[['補助事項或用途','補助計畫案總經費及分攤情形-合計']].groupby(['補助事項或用途'])['補助計畫案總經費及分攤情形-合計'].sum().reset_index()
```

```python
hello['季度'] = 3
hello
```

```python
bye = df4[['補助事項或用途','補助計畫案總經費及分攤情形-合計']].groupby(['補助事項或用途'])['補助計畫案總經費及分攤情形-合計'].sum().reset_index()
```

```python
bye['季度'] = 4
bye
```
